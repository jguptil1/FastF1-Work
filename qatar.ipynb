### Imports

```python
import pandas as pd
import fastf1
```

### API

```python
session = fastf1.get_session(2023, 'Qatar', 'R')
session.load()
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', None)
```

### Session Load

```python
a = session.laps
race_laps = pd.DataFrame(a)
race_laps.head()
```

### Data Transformation of Various Times

```python
race_laps['LapTime'] = pd.to_timedelta(race_laps['LapTime'])
race_laps['minutes'] = race_laps['LapTime'].dt.components.minutes
race_laps['seconds'] = race_laps['LapTime'].dt.components.seconds
race_laps['milliseconds'] = race_laps['LapTime'].dt.components.milliseconds
race_laps['TotalSeconds'] = race_laps['LapTime'].dt.total_seconds()
race_laps['Sector1TimeTotal'] = race_laps['Sector1Time'].dt.total_seconds()
race_laps['Sector2TimeTotal'] = race_laps['Sector2Time'].dt.total_seconds()
race_laps['Sector3TimeTotal'] = race_laps['Sector3Time'].dt.total_seconds()
race_laps[['LapTime','minutes', 'seconds', 'milliseconds', 'TotalSeconds']]
race_laps.head()
```

### Data Cleaning and Subsetting

```python
race_laps.columns
```

```python
race_laps_sub = race_laps[['Time', 'Driver', 'DriverNumber', 'LapTime', 'LapNumber', 'TotalSeconds', 'TyreLife', 'Position', 'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST','Sector1TimeTotal', 'Sector2TimeTotal',
       'Sector3TimeTotal']]
race_laps_sub.head()
race_laps_sub.isna().sum()
```

I need to find every lap that has a missing value. This might be related to a pit stop of some sort

```python
# Find records in ColumnA where ColumnB has missing values
result = race_laps_sub[race_laps_sub['LapNumber'].isna()][['LapTime']]
                                                           
                                                           
#'TotalSeconds', 'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST','Sector1TimeTotal', 'Sector2TimeTotal', 'Sector3TimeTotal']]
result
```
